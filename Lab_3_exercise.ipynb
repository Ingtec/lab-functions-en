## 2. Now write a function that returns the largest element on a list

```python
def greatest(lst):
    """

    Args:
      lst:

    Returns:

    """
    max_value = lst[0]
    for num in lst:
        if num > max_value:
            max_value = num
    return max_value
```

## 6. Write a function that returns the factorial of a number.

```python
def factorial(n):
    """

    Args:
      n:

    Returns:

    """
    result = 1  # Initialize result as 1 since factorial multiplication starts with 1
    for i in range(1, n + 1):  # Iterate from 1 to n inclusive
        result *= i  # Multiply current number with result
    return result  # Return the final factorial value
```

```python
#factorial formula
#n! = n * ( n - 1 ) *...*1

# This code defines a function called "factorial" which takes an input "n". The function uses a for loop to iterate through the range of numbers
# from 1 to n+1. For each number in that range, it multiplies the current value of x by the number in the range. At the end of the loop,
# the function returns the final value of x, which will be the factorial of the input number "n".

# The Factorial of a positive integer n is the product of all positive integers less than or equal to n.
# For example, the factorial of 6 (written "6!") is 6 * 5 * 4 * 3 * 2 * 1 = 720.

# So this function takes an input of any positive integer, and returns the factorial of that number.
```

```python
# This will test your function
test_factorial(factorial)
```

## 7. Write a function that takes a list and returns a list of the unique values.

`NOTE: You cannot use set. ðŸ¤”`

```python
def unique(lst_un):
    """

    Args:
      lst_un:

    Returns:

    """
    # Initialize an empty list to store unique values
    unique_list = []
    # Iterate through each element in the input list
    for item in lst_un:
        # Add the item to the unique list if it's not already there
        if item not in unique_list:
            unique_list.append(item)
    # Return the list of unique values
    return unique_list
```

## 8. Write a function that returns the mode of a list, i.e.: the element that appears the most times.
`NOTE: You should not use count... ðŸ§`

```python
def mode_counter(arr):
    """

    Args:
      arr:

    Returns:

    """
    # Create a dictionary to store the frequency of each element
    frequency = {}
    for item in arr:
        if item in frequency:
            frequency[item] += 1
        else:
            frequency[item] = 1

    # Find the element with the highest frequency
    max_count = 0
    mode = None
    for key, value in frequency.items():
        if value > max_count:
            max_count = value
            mode = key

    return mode
```

## 9. Write a function that calculates the standard deviation of a list.
`NOTE: Do not use any libraries or already built functions. ðŸ˜‰`

```python
def st_dev(list_sd):
    # Step 1: Calculate the mean of the list
    total = 0
    n = len(list_sd)  # Number of elements in the list
    for num in list_sd:
        total += num
    mean = total / n

    # Step 2: Calculate the sum of squared differences from the mean
    variance_sum = 0
    for num in list_sd:
        variance_sum += (num - mean) ** 2

    # Step 3: Calculate variance by dividing the sum by n-1 (sample variance)
    variance = variance_sum / (n - 1)

    # Step 4: Take the square root of the variance to get the standard deviation
    std_dev = variance ** 0.5

    return std_dev
```

```python
# This will test your function
test_stdev(st_dev)
```

## 10. Write a function to check if a string is a pangram, i.e.: if it contains all the letters of the alphabet at least once. Mind that the strings may contain characters that are not letters.

```python
def pangram(string):
    # Step 1: Initialize an empty set to store unique alphabet letters found in the string
    letters_found = set()

    # Step 2: Iterate through each character in the string
    for char in string:
        # Step 3: Convert the character to lowercase and check if it's a letter
        if 'a' <= char.lower() <= 'z':
            letters_found.add(char.lower())  # Add the letter to the set

    # Step 4: Check if the length of the set is 26 (all letters of the alphabet are present)
    return len(letters_found) == 26
```

```python
# This will test your function
test_pangram(pangram)
```

## 11. Write a function that receives a string of comma separated words and returns a string of comma separated words sorted alphabetically.

`NOTE: You may use sorted but not split and definitely no join! ðŸ¤ª`

```python
def sort_alpha(string):
    # Step 1: Parse the string manually and store words in a list
    words = []
    current_word = ""
    for char in string:
        # Step 2: If the character is a comma, finalize the current word
        if char == ',':
            words.append(current_word)
            current_word = ""
        else:
            current_word += char
    words.append(current_word)  # Append the last word

    # Step 3: Sort the words using the sorted function
    sorted_words = sorted(words)

    # Step 4: Manually reassemble the sorted words into a single string
    result = ""
    for word in sorted_words:
        result += word + ','  # Add a comma after each word
    return result[:-1]  # Remove the trailing comma
```

```python
# This will test your function
test_alpha(sort_alpha)
```

## 12. Write a function to check if a given password is strong (at least 8 characters, at least one lower case, at least one upper case, at least one number and at least one special character). It should output True if strong and False if not.

```python
def check_pass(password):
    # Step 1: Initialize variables to check the requirements
    has_lower = False
    has_upper = False
    has_digit = False
    has_special = False

    # Step 2: Define a string of special characters
    special_characters = "!@#$%^&*()-_=+[]{}|;:'\",.<>?/"

    # Step 3: Check the length of the password
    if len(password) < 8:
        return False  # Password is not strong if it's less than 8 characters

    # Step 4: Iterate through each character in the password
    for char in password:
        if char.islower():
            has_lower = True
        elif char.isupper():
            has_upper = True
        elif char.isdigit():
            has_digit = True
        elif char in special_characters:
            has_special = True

        # Step 5: If all conditions are met, no need to continue checking
        if has_lower and has_upper and has_digit and has_special:
            return True

    # Step 6: Return False if any condition is not met
    return False
```

```python
# This will test your function
test_pass(check_pass)
```
